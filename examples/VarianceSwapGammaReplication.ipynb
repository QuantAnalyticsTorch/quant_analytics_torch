```python
import torch
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def black_call(s, k, dt, v, r):
    n = torch.distributions.Normal(0, 1).cdf
    sdt = v * torch.sqrt(dt)
    d1 = (torch.log(s / k) + (r + v * v / 2) * dt) / sdt
    d2 = d1 - sdt
    return s * n(d1) - k * torch.exp(-r * dt) * n(d2)


def black_put(s, k, dt, v, r):
    n = torch.distributions.Normal(0, 1).cdf
    sdt = v * torch.sqrt(dt)
    d1 = (torch.log(s / k) + (r + v * v / 2) * dt) / sdt
    d2 = d1 - sdt
    return -s * n(-d1) + k * torch.exp(-r * dt) * n(-d2)
```

```python
s = torch.tensor(100.,requires_grad=True)
k = torch.tensor(100.)
dt = torch.tensor(1.)
v = torch.tensor(0.2)
r = torch.tensor(0.)
```

```python
result = black_call(s,k,dt,v,r)
```

```python
dx = torch.autograd.grad(result, s, create_graph=True, retain_graph=True, allow_unused=True)
```

```python
ddx = torch.autograd.grad(dx, s, create_graph=True, retain_graph=True, allow_unused=True)
```

```python
import numpy as np
import pandas as pd
```

```python
spots = np.linspace(50,130,17)
strikes = np.linspace(10,190,19)
#strikes = [100]
```

```python
df = pd.DataFrame()

for it in spots:
    s = torch.tensor(it,requires_grad=True)
    result = torch.tensor(0.)
    for jt in strikes:
        if jt < 100:
            result = result + black_put(s,jt,dt,v,r) / jt / jt
        else:
            result = result + black_call(s,jt,dt,v,r) / jt / jt
    dx = torch.autograd.grad(result, s, create_graph=True, retain_graph=True, allow_unused=True)
    ddx = torch.autograd.grad(dx, s, create_graph=True, retain_graph=True, allow_unused=True)
    
    row = { 'spot' : it, 'value' : result.detach().numpy(), 'cash-gamma' : it * it * ddx[0].detach().numpy(), 'cash-delta' : it * dx[0].detach().numpy() }
    df = df.append(row,ignore_index=True)
df
```

```python
plt.plot(df['spot'], df['cash-gamma'])
plt.show()
```

```python
plt.plot(df['spot'], df['cash-delta'])
plt.show()
```

```python
df_cash_gamma = pd.DataFrame()

for it in spots:
    s = torch.tensor(it,requires_grad=True)
    row = { 'spot' : it }
    result = torch.tensor(0.)
    for jt in strikes:
        if jt < 100:
            result = black_put(s,jt,dt,v,r) / jt / jt
        else:
            result = black_call(s,jt,dt,v,r) / jt / jt
        dx = torch.autograd.grad(result, s, create_graph=True, retain_graph=True, allow_unused=True)
        ddx = torch.autograd.grad(dx, s, create_graph=True, retain_graph=True, allow_unused=True)
    
        row['cash-gamma' + str(int(jt))] = it * it * ddx[0].detach().numpy()

    df_cash_gamma = df_cash_gamma.append(row,ignore_index=True)
df_cash_gamma
```
